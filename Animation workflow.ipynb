```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
x = np.linspace(0, 2 * np.pi, 100)

fig = plt.figure()
#get the line
lines = plt.plot([])
line = lines[0]

#other setup
plt.xlim(0, 2*np.pi)
plt.ylim(-1.1, 1.1)

#animate functions
def animate(frame):
    y = np.sin(x + 2*np.pi * frame/100)
    line.set_data((x, y))

#call this to animate 
anim = FuncAnimation(fig, animate, frames =100, interval = 16.6)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html) #draws the display
plt.close()
```

```python
y, x = np.ogrid[-5:5:100j, -5:5:100j]
#play with the limits of a and b
a = np.linspace(-10, 10, 300) 
b = np.linspace(-10, 10, 300)

fig = plt.figure()
#get the line
lines = plt.plot([])
line = lines[0]

#other setup
plt.xlim(-5, 5)
plt.ylim(-5, 5)

#animate functions
def animate(frame):
    fig.clear()
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a[frame] - b[frame], [0])

#call this to animate 
anim = FuncAnimation(fig, animate, frames =300, interval = 16.6)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html) #draws the display
plt.close()
```
